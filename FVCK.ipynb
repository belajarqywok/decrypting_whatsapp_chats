<h3>Message Store Baru <b>(msgstore.new.db)</b></h3>

```python
cursor.execute(
    """
    SELECT datetime(timestamp / 1000, 'unixepoch') AS readable_time, text_data
    FROM message
    WHERE text_data LIKE '%cintia%';
    """
)

tables = cursor.fetchall()
for row in tables: print(list(row))
```

```python
for row in tables:
  data = list(row)
  print(f"[ Date ]: {data[0]}")
  print(f"[ Message ]: {data[1]}")

  separator = "\n" + "-" * 53
  print(f"{separator * 3}\n")
```
